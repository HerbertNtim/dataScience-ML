### Using object-oriented interface over pyplot interface

```python
# 1st Method
fig = plt.figure() # Creates figure
ax = fig.add_subplot() # Add axes
plt.show();
```

```python
# 2nd Method
fig = plt.figure()
ax = fig.add_axes([1, 1, 1, 1])
plt.plot(x, y)
plt.show();
```

```python
# 3rd Method (Recommended)
fig, ax = plt.subplots()
ax.plot(x, y);
```

```python
fig, ax = plt.subplots()
ax.plot(x, [50, 100, 200, 250]);
```

### Matplotlib Example Workflow

```python
# 0. import matplotlib and get it ready for plotting in Jupyter Notebook
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare Data
x = [1, 2, 3, 4, 5]
y = [11, 22, 33, 44, 55]

# 2. Setup plot and give a figure size
fig, ax = plt.subplots(figsize=(10,10))

# 3. Plot data 
ax.plot(x, y)

# 4. Customize plot
ax.set(title="Simple Matplotlib plot", xlabel="x-axis", ylabel="y-axis")

# 5. Save & show (you save the whole figure)
fig.savefig("images/simple-plot2.png")
```

### Making figures With Numpy Arrays

```python
# import numpy 
import numpy as np
```

```python
# Create some data using linspace and plot a line plot
x = np.linspace(0, 10, 100)

fig, ax = plt.subplots()
ax.plot(x, x**2);
```

```python
# Using the same x data to plot a scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x));
```

```python
# Another sin wave scatter plot 
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x));
```

```python
# Make a plot from dictionary
nut_butter = { "Almond butter": 8, "Peanut butter": 10, "Cashew butter": 12 }
fig, ax = plt.subplots()
ax.bar(nut_butter.keys(), nut_butter.values())
ax.set(title="Nuts Butter", ylabel="Price ($)");
```

```python
# Horizontal bar graph
fig, ax = plt.subplots()
ax.barh(nut_butter.keys(), nut_butter.values());
```
