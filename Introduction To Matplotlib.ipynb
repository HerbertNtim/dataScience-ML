### Matplotlib Example Workflow

```python
# 0. import matplotlib and get it ready for plotting in Jupyter Notebook
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare Data
x = [1, 2, 3, 4, 5]
y = [11, 22, 33, 44, 55]

# 2. Setup plot and give a figure size
fig, ax = plt.subplots(figsize=(10,10))

# 3. Plot data 
ax.plot(x, y)

# 4. Customize plot
ax.set(title="Simple Matplotlib plot", xlabel="x-axis", ylabel="y-axis")

# 5. Save & show (you save the whole figure)
fig.savefig("images/simple-plot2.png")
```

## Two Options for Subplots

```python
# Subplot Option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(10, 5))
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter.keys(), nut_butter.values());
ax4.hist(np.random.randn(1000));
```

```python
# Subplots Option 2
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10, 5));
ax[0, 0].plot(x, x/2);
ax[0, 1].scatter(np.random.random(10), np.random.random(10));
ax[1, 0].bar(nut_butter.keys(), nut_butter.values());
ax[1, 1].hist(np.random.randn(1000));
```

## Plotting from Pandas DataFrames

```python
# Make data
import pandas as pd
car_sales = pd.read_csv("car-sales.csv")
car_sales
```

```python
# Formatting Car Price Data

car_sales["Price"] = car_sales["Price"].str.replace('$', '')
car_sales["Price"] = car_sales["Price"].str.replace(',', '')
car_sales["Price"] = car_sales["Price"].str.replace('.', '')
car_sales
```

```python
# Removing the last two zeros from the price column
car_sales['Price'] =  car_sales['Price'].str[:-2]
car_sales
```

```python
# Adding Sale date to car car sales data
car_sales["Sale Date"] = pd.date_range("1/1/2026", periods = len(car_sales))
car_sales
```

```python
# Add Total Sales using the prices column
car_sales["Total Sales"] = car_sales["Price"].astype(int).cumsum()
car_sales
```

```python
car_sales.plot(x = "Sale Date", y = "Total Sales");
```

```python
car_sales.plot(x="Odometer (KM)", y="Price", kind="scatter");
```

```python
# Plotting data using different data set
heart_disease = pd.read_csv('heart-disease.csv')
heart_disease.head()
```

```python
# Trying the difference between the pyplot and matplotlib OO methods
over_50 = heart_disease[heart_disease['age'] > 50]
over_50.head()
```

```python
# Pyplot Method
over_50.plot(kind='scatter', x='age', y='chol', c='target');
```

```python
# OO Method from matplotlib
fig, ax = plt.subplots(figsize=(10, 6))
over_50.plot(kind="scatter", x='age', y='chol', c='target', ax=ax);
# ax.set_xlim([45, 100])
```

```python
# OO Method from Scratch
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the data
scatter = ax.scatter(x=over_50['age'], y=over_50['chol'], c=over_50['target']);

# Customize the plot
ax.set(title="Heart Disese and Cholesterol Levels", xlabel='Age', ylabel='Cholesterol');

# Adding horizontal line to the scatter plot
ax.axhline(y=over_50['chol'].mean(), color='b', linestyle='--');

# Add legend
ax.legend(*scatter.legend_elements(), title = 'Target');
```

```python
# Subplots of age, cholesterol, thalach
fig, (ax0, ax1) = plt.subplots(nrows=2, ncols=1, figsize=(10, 10))

# Add data to ax0
ax0_scatter = ax0.scatter(x=over_50['age'], y=over_50['chol'], c=over_50['target']);

# Customize ax0
ax0.set(title="Heart Disease with cholesterol and Age", xlabel="Age", ylabel="Cholesterol");

# Add a legend to ax0
ax0.legend(*ax0_scatter.legend_elements(), title='Target');

# Add a mean line
ax0.axhline(y=over_50['chol'].mean(), linestyle='--');

# Add data to ax1
ax1_scatter = ax1.scatter(x=over_50['age'], y=over_50['thalach'], c=over_50['target']);

# Customize ax1
ax1.set(title="Heart Disease with Thalach and Age", xlabel="Age", ylabel="Thalach");

# Add a legend to ax1
ax1.legend(*ax1_scatter.legend_elements(), title='Target');

# Add a mean line
ax1.axhline(y=over_50['thalach'].mean(), linestyle='--');

# Add title to the figure
fig.suptitle('Heart Disease Analysis', fontsize=16, fontweight='bold');
```
