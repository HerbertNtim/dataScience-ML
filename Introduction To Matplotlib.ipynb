```python
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]
plt.plot(x, y);
```

### Using object-oriented interface over pyplot interface

```python
# 1st Method
fig = plt.figure() # Creates figure
ax = fig.add_subplot() # Add axes
plt.show();
```

```python
# 2nd Method
fig = plt.figure()
ax = fig.add_axes([1, 1, 1, 1])
plt.plot(x, y)
plt.show();
```

```python
# 3rd Method (Recommended)
fig, ax = plt.subplots()
ax.plot(x, y);
```

```python
fig, ax = plt.subplots()
ax.plot(x, [50, 100, 200, 250]);
```
